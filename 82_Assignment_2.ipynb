```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d1 = 2  # last digit of student ID
d2 = 8  # second last digit of student ID

A = np.array([[d1 + 2, d2 + 1],
              [2 * d1, d2 + 2]])


print("Matrix A:")
print(A)

print("Shape of A:", A.shape)
```

```python
det_A = np.linalg.det(A)
print("Determinant of A:", det_A)

rank_A = np.linalg.matrix_rank(A)
print("Rank of A:", rank_A)
```

```python

eigvals_A = np.linalg.eigvals(A)
print("Eigenvalues of A:", eigvals_A)

if det_A != 0:
    inv_A = np.linalg.inv(A)
    print("Inverse of A:")
    print(inv_A)
else:
    print("Matrix A is not invertible.")
```

```python
# Add +1 to element A[0,0]
B = A.copy()
B[0, 0] += 1

print("\nMatrix B (A[0,0] increased by +1):")
print(B)
```

```python

print("Shape of B:", B.shape)

det_B = np.linalg.det(B)
print("Determinant of B:", det_B)

rank_B = np.linalg.matrix_rank(B)
print("Rank of B:", rank_B)

eigvals_B = np.linalg.eigvals(B)
print("Eigenvalues of B:", eigvals_B)

if det_B != 0:
    inv_B = np.linalg.inv(B)
    print("Inverse of B:")
    print(inv_B)
else:
    print("Matrix B is not invertible.")
```
