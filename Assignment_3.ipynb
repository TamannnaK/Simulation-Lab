```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Part 1 — Data Generation

# Regular days: Uniform distribution between 2 and 12 minutes
regular_wait_times = np.random.uniform(2, 12, 3000)

# Busy days: Normal distribution (mean=8, std=2)
busy_wait_times = np.random.normal(8, 2, 3000)

# 3. Replace negative wait times with 0
busy_wait_times = np.clip(busy_wait_times, 0, None)
```

```python
# Part 2 — Visualization


plt.figure()
plt.hist(regular_wait_times, bins=30, alpha=0.7)
plt.hist(busy_wait_times, bins=30, alpha=0.7)
plt.xlabel("Wait Time (minutes)")
plt.ylabel("Frequency")
plt.title("Customer Wait Time Distributions")
plt.show()
```

```python
# Part 3 — Statistical Analysis

def print_statistics(data, label):
    print(f"\n{label} Statistics")
    print(f"Mean: {np.mean(data):.2f}")
    print(f"Median: {np.median(data):.2f}")
    print(f"Standard Deviation: {np.std(data):.2f}")
    print(f"Maximum: {np.max(data):.2f}")

print_statistics(regular_wait_times, "Regular Day")
print_statistics(busy_wait_times, "Busy Day")
```
